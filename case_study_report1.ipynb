```python
#delete unamed column
df = df.drop(columns=['Unnamed: 9'])
```

```python
edu_map = {
    "0":np.nan,
    "Not sure":np.nan,
    "Ninguno": 0,
    "Incomplete primary": 1,
    "Complete primary": 2,
    "Incomplete Secundary": 3,
    "Complete Secundary": 4,
    "Incomplete technical or technological": 5,
    "Complete technique or technology": 6,
    "Incomplete Professional Education": 7,
    "Complete professional education": 8,
    "Postgraduate education": 9,
}

#define an education map, so that the 12 levels of educational stage
#could be trnasformed to standard scale for comparision

df["edu_father_scale"] = df["EDU_FATHER"].map(edu_map)
df["edu_mother_scale"] = df["EDU_MOTHER"].map(edu_map)
df['edu_average'] = df[['edu_father_scale', 'edu_mother_scale']].mean(axis=1,skipna=True)
df['edu_average'].head()

#take the average level of parents as family average education level
```

```python
#separate student performance into two stages
first_stage = df[['MAT_S11','CR_S11','CC_S11','BIO_S11','ENG_S11']]
second_stage = df[['CR_PRO','QR_PRO','CC_PRO','WC_PRO','ENG_PRO']]
df['first_average'] = first_stage.mean(axis=1, skipna=True)
df['second_average'] = second_stage.mean(axis=1, skipna=True)
df['personalgrowth'] = df['second_average'] - df['first_average']
display(df['personalgrowth'].head())
```

```python
#stastical properties
print(df[['edu_average','first_average','second_average','personalgrowth']].describe())
```

```python
# see the correlation between family education and student's performance in two stages
print(df[['edu_average','first_average','second_average','personalgrowth']].corr())
```

```python
# see whether how parents impact children's across gender more precisely
corr_gender = df.groupby('GENDER')[['edu_father_scale','edu_mother_scale','personalgrowth']].corr().round(2)
print(corr_gender)
```

```python
# how could family educational level impact student's growth
print(df[['edu_average','personalgrowth']].describe())
```

```python
# Capacity of reading and expression
df['lang_first'] = df[['ENG_S11', 'CR_S11']].mean(axis=1, skipna=True)
df['lang_second'] = df[['ENG_PRO', 'CR_PRO']].mean(axis=1, skipna=True)
df['lang_diff'] = df['lang_second'] - df['lang_first']

# Capacity of mathematical reasoning
df['math_first'] = df[['MAT_S11', 'BIO_S11']].mean(axis=1, skipna=True)
df['math_second'] = df['CR_PRO']
df['math_diff'] = df['math_second'] - df['math_first']

# Capacity of social and communication
df['social_first'] = df['CC_S11']
df['social_second'] = df[['CC_PRO', 'WC_PRO']].mean(axis=1, skipna=True)
df['social_diff'] = df['social_second'] - df['social_first']
```

```python
capacity_table = df[['lang_first','lang_second',
            'math_first','math_second',
            'social_first','social_second','G_SC']].corr().round(2)
capacity_table
```
