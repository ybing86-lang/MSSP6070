```python
# 2.d
ring_of_fire_locations = ["Bolivia", "Chile", "Ecuador", "Peru", "Costa Rica", "Guatemala", "Mexico", "Japan", "Philippines", "Indonesia",
              "New Zealand","Antarctic", "Canada", "Fiji", "Alaska", "Washington", "California"]

df["Ring_of_Fire"] = df["parsed_place"].apply(lambda x: any(loc in x for loc in ring_of_fire_locations) if pd.notna(x) else False)

# The code for filtering Mexico and excluding New Mexico was part of a previous task
# I will keep it as is, but it is not directly related to the Ring of Fire calculation
# mask = df["parsed_place"].str.contains("Mexico", case=False, na=False)&(df["parsed_place"]!="New Mexico")
# result = df[mask]
# list(ring_of_fire_locations)

df.head()
```

```python
# 2.e
ring_of_fire_count = df['Ring_of_Fire'].sum()
outside_ring_count = len(df) - ring_of_fire_count
total_count = len(df)
ring_percentage = (ring_of_fire_count/total_count)*100
outside_percentage = (outside_ring_count/total_count)*100

print(f"Number of earthquakes in the Ring of Fire: {ring_of_fire_count}")
print(f"Number of earthquakes outside the Ring of Fire: {outside_ring_count}")
print(f"Percentage of earthquakes in the Ring of Fire: {ring_percentage:.2f}%")
print(f"Percentage of earthquakes outside the Ring of Fire: {outside_percentage:.2f}%")
```

```python
# 2.f
ring_with_tsunami = df[(df["Ring_of_Fire"] == True) & (df["tsunami"] == 1)]
tsunami_count = len(ring_with_tsunami)
print("Tsunami count in Ring of Fire:", tsunami_count)
```
