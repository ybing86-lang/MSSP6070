```python
df = pd.read_csv('/content/Student Performance Data.csv')
df.head()
```

```python
df.columns
```

```python
#calculate total points
df['score_total'] = df.apply(lambda row: row['math score'] + row['reading score'] + row['writing score'], axis=1)

#calculate average points
df['average_score'] = df.apply(lambda row: (row['math score'] + row['reading score'] + row['writing score']) / 3, axis=1)

#evaluate student level
df['student_class'] = df['average_score'].apply(
    lambda x: 'excellent-A' if x >= 90 else
              'good-B' if 80 <= x < 90 else
              'pass-C' if 60 <= x < 80 else
              'fail-D'
)

#evaluate whether one student passes the test
df['pass_status'] = df.apply(lambda row: 'pass' if row['math score'] >= 60 and row['reading score'] >= 60 and row['writing score'] >= 60 else 'fail', axis=1)
```
