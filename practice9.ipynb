```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
url = "https://guides.library.upenn.edu/americannewspapers/present"
```

```python
res = requests.get(url)
res.raise_for_status()
```

```python
soup = BeautifulSoup(res.text, "html.parser")
```

```python
images = soup.find_all("img")
```

```python
folder = "upenn_images"
```

```python
for img in images:
    src = img.get("src")
    if not src:
        continue
    if src.startswith("//"):
        src = "https:" + src
    elif src.startswith("/"):
        src = "https://guides.library.upenn.edu" + src
```

```python
img_data = requests.get(src).content
```

```python
file_name = os.path.basename(src)
```

```python
os.makedirs(folder, exist_ok=True)
with open(os.path.join(folder, file_name), "wb") as f:
        f.write(img_data)
```
