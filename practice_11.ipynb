```python
geojson_file = "/content/california-congressional-districts.geojson"
with open(geojson_file) as file:
    geojson_data = json.load(file)
```

```python
fig, ax = plt.subplots(figsize=(8, 12))
for district in geojson_data["features"]:
    district_shape = shape(district["geometry"])
    if isinstance(district_shape, Polygon):
        x, y = district_shape.exterior.xy
        ax.fill(x, y, alpha=0.5)
    elif isinstance(district_shape, MultiPolygon):
        for poly in district_shape.geoms:
            x, y = poly.exterior.xy
            ax.fill(x, y, alpha=0.5)

ax.set_title("California Congressional Districts")
plt.show()
```
