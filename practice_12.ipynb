```python
import io
df = pd.read_csv(io.BytesIO(uploaded['Student Performance Data (2).csv']), encoding='latin-1')
```

```python
df.head()
```

```python
males = df[df["gender"] == "male"]
male_group = males[males["test preparation course"] == "completed"]["total score"]
others = df[df["test preparation course"] != "completed"]["total score"]
```

```python
t,p=stats.ttest_ind(male_group, others, equal_var=False, nan_policy="omit")
float(t),float(p)
```

```python
male_writing = df[df["gender"] == "male"]["writing score"]
female_writing = df[df["gender"] == "female"]["writing score"]

t,p=stats.ttest_ind(male_writing, female_writing, equal_var=False, nan_policy="omit")
float(t),float(p)
```
