```python
import zipfile

zip_path = "LLCP2022XPT.zip"
extract_dir = "/content/brfss2022"

with zipfile.ZipFile(zip_path, 'r') as z:
    z.extractall(extract_dir)
```

```python
file_name = "LLCP2022XPT.zip"
df_raw = pd.read_sas(
    file_name,
    format="xport",
    encoding="latin1",
    compression="zip"
)
df_raw.shape
```

```python
cols_report4 = [
    "MENTHLTH", "_MENT14D",                      # Psychological status
    "INCOME3", "_INCOMG1", "EMPLOY1", "EDUCA",   # Economic status indicators
    "_HLTHPLN", "PERSDOC3", "MEDCOST1",           # Healthcare access variables
    "_STATE", "SEXVAR", "_AGEG5YR",             # Basic demographic controls
    "_RACEGR4", "MARITAL", "GENHLTH"              # Additional demographic/health controls
]
df_subset = df_raw[cols_report4].copy()
df_subset.head()
```

```python
## cleaning data
missing_codes = [7, 9, 77, 99, 777, 999]
df_clean = df_subset[cols_report4].copy()

for col in df_clean.columns:
    df_clean[col] = pd.to_numeric(df_clean[col], errors="ignore")
df_clean = df_clean.replace(missing_codes, np.nan)
df_clean["MENTHLTH"] = df_clean["MENTHLTH"].replace(88, 0)

df_clean.shape
```
